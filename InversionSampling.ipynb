```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline

def p2c(rad,thet):#polar to cartesian
    x = rad * np.cos(thet)
    y = rad * np.sin(thet)
    return (x,y)

N = 50000

theta = np.random.rand(N) * 2 * np.pi #samples in [0,2pi]
r_sq = np.random.rand(N) #squared samples in [0,1]
r = np.sqrt(r_sq)

xCoords, yCoords = p2c(r,theta)

plt.figure(figsize=(9,9))
plt.plot(xCoords,yCoords,'r.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sampling from Unit Cube Using Inversion Sampling')
plt.show()
```
