```python
import numpy as np
import matplotlib.pylab as plt
import matplotlib.patches as mptc
%matplotlib inline

def rot(xp, yp, thet):#rotate (xp,yp) by thet angles
    x = xp * np.cos(thet) - yp * np.sin(thet)
    y = xp * np.sin(thet) + yp * np.cos(thet)
    return (x,y)

def pNorm(p, x, y):
    xLst = []
    yLst = []
    for i in range(len(x)):
        if np.power(np.fabs(x[i]),p) + np.power(np.fabs(y[i]),p) <= 1:
            xLst.append(x[i])
            yLst.append(y[i])
    return(xLst,yLst)

N = 500000
sq2 = np.sqrt(2)

x = np.random.rand(N) * sq2 - sq2/2
y = np.random.rand(N) * sq2 - sq2/2

xCoords, yCoords = rot(x, y, (np.pi/4))

x_p1, y_p1 = pNorm(0.7, xCoords, yCoords)

ar_1 = (float)(len(x_p1))/(float)(len(xCoords))#acceptance rate for pnorm(1.5)

plt.figure(figsize=(10,10))
plt.plot(xCoords,yCoords,'b.',label='Proposal: rotated square')
plt.plot(x_p1,y_p1,'r.', label='p=0,7')
plt.title('p=0.7, acceptanceRate = ' + str(ar_1))
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right', fontsize='medium')

plt.show()
```
