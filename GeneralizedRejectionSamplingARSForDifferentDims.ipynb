```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

def pNorm(data):
    (m,n) = np.shape(data) # mxn matrix (m = N; n = dim actually)
    acceptedData = data #initialization
    count = 0
    for i in range(m):
        pTemp = 0
        for j in range(n):
            pTemp += np.power(np.fabs(data[i,j]),2)
        if  pTemp <= 1: # |x|^dim + |y|^dim ... <= 1^dim
            acceptedData[count,:] = data[i,:]
            count += 1
    acceptedData = acceptedData[:count] #discard rest
    return(acceptedData,count)

ars = np.zeros(11)
for d in range(1,11):
    N = 100000 #no of samples
    dim = d #<dim>-dimensional space
    
    data = np.zeros((N,dim))
    for i in range(dim): #(n x dim) uniform numbers from [-1,1]
        data[:,i] = np.random.rand(N) * 2 - 1
    
    accepted, count = pNorm(data)
    ars[d] = (float)(count) / len(data)
    
plt.figure(figsize=(7,7))
plt.plot(ars,'r-x')
plt.title('Acceptance Rates for Different Dimensions in Rejection Sampling')
plt.xlabel('dimension')
plt.ylabel('acceptance rate')
plt.xlim([1,10])
plt.show()

print ars
```
