```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N = 4200

theta = np.random.rand(N)*2*np.pi
r = np.sqrt((np.random.rand(N))) #This part uses inversion.  u=F(r)=r^2  => F−1(u)=r=sqrt(u) 
x = r*np.cos(theta)
y = r*np.sin(theta)

plt.axis('equal')
plt.plot(x, y, '.m')
```

```python
import numpy as np
import matplotlib.pyplot as plt


N = 4200
p=0.7
accepted=0

theta = np.random.rand(N)*2*np.pi
r = np.sqrt((np.random.rand(N))) #This part uses inversion.  u=F(r)=r^2  => F−1(u)=r=sqrt(u) 
x = r*np.cos(theta)
y = r*np.sin(theta)

ax=[] #accepted x
ay=[] #accepted y


for i in range(N):
    if((abs(x[i])**p)+(abs(y[i])**p) <=1):
        ax.append(x[i])
        ay.append(y[i])
        accepted += 1
    
        
print("The acceptance rate is: ", accepted/N)

plt.axis('equal')
plt.plot(x, y, '.m')
plt.plot(ax, ay, '.c')
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

N=4200
p=0.7

#we draw from points uniformly from a square, which has side length square root 2 (to make the diagonal 1)

x=np.random.rand(N)*np.sqrt(2)
y=np.random.rand(N)*np.sqrt(2)

#shift it to place its center to origin

x-=2**0.5/2
y-=2**0.5/2

#rotate it 45 degrees anti-clockwise

rx=(np.sqrt(2)*x-np.sqrt(2)*y)/2
ry=(np.sqrt(2)*x+np.sqrt(2)*y)/2

#rejection sampling

ax=[]
ay=[]
  
accepted=((abs(rx)**p)+(abs(ry)**p) <=1)
ax=rx[accepted]
ay=ry[accepted]


plt.plot(rx,ry,'.m')
plt.plot(ax,ay,'.c')
plt.axis('equal')
```
