```python
import numpy as np
import matplotlib.pyplot as plt


N = 4200
p=0.7
accepted=0

theta = np.random.rand(N)*2*np.pi
r = np.sqrt((np.random.rand(N))) #This part uses inversion.  u=F(r)=r^2  => F−1(u)=r=sqrt(u) 
x = r*np.cos(theta)
y = r*np.sin(theta)

ax=[] #accepted x
ay=[] #accepted y


for i in range(N):
    if((abs(x[i])**p)+(abs(y[i])**p) <=1):
        ax.append(x[i])
        ay.append(y[i])
        accepted += 1
    
        
print("The acceptance rate is: ", accepted/N)

plt.axis('equal')
plt.plot(x, y, '.m')
plt.plot(ax, ay, '.c')
```
