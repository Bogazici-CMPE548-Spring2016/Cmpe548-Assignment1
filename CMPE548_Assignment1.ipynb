###1. Uniform samples from a circular region###

If a is chosen with probability density $f(a)$, then the integrated probability
up to point a, $F(a)$, is itself a random variable which will occur with uniform probability
density on [0, 1]. 

If x can take on any value, we can then find a unique x chosen from the p.d.f. for a given u if we set 

$u = F(x)$

provided we can find an inverse of F, defined by

$x = F^{−1}(u)$

For the problem,

$u = F(r) = \int_{0}^{r} 2x dx = r^2$

and, 

$x = F^{-1}(u) = \sqrt u$

Thus, taking a value on uniform probability density on $[0, 1]$ and square root of it gives sample from $p(r)$.   

```python
%matplotlib inline

import random
import numpy as np 
import matplotlib.pyplot as plt

N=5000

r = [np.sqrt(random.random()) for i in range(N)]
theta = [random.random()*2*np.pi for i in range(N)]

x = np.multiply(r,np.cos(theta)) # x = rcos(theta)
y = np.multiply(r,np.sin(theta)) # y = rsin(theta)
    
plt.axis('equal')
plt.plot(x, y, 'k.',label = '')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

### 2. Rejection Sampling ###

```python
%matplotlib inline

N = 5000.0

for i in [0.7, 1.5]:
    x_list = []
    y_list = []
    count = 0.0
    for j in range(int(N)):
        r = np.sqrt(random.random())
        theta = random.random()*2*np.pi
        x = np.multiply(r,np.cos(theta))
        y = np.multiply(r,np.sin(theta))
        check = np.power(np.power(np.absolute(x),i)+np.power(np.absolute(y),i),np.divide(1,i))
        if check <= 1:
            count+=1.0
            x_list.append(np.multiply(r,np.cos(theta))) # x = rcos(theta)
            y_list.append(np.multiply(r,np.sin(theta))) # y = rsin(theta)
    plt.axis('equal')
    plt.plot(x_list, y_list, 'k.',label = 'p = '+str(i))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()
    print 'Acceptance Rate = ',np.divide(count,N)
    
        
```

Since acceptance rate for $p = 0.7$ is 0.4052, it can be improved by selecting proposal as 1 norm unit ball. So as to take uniform samples from 1 norm unit ball, we can generate x and y coordinates unifromly from $[0,1]$ which generates a square and rotate it.

```python
%matplotlib inline


N=2500

x = [random.random()*np.sqrt(2)-np.divide(1,np.sqrt(2)) for i in range(N)]
y = [random.random()*np.sqrt(2)-np.divide(1,np.sqrt(2)) for i in range(N)]

rotated_x = np.multiply(x,np.cos(np.divide(np.pi,4)))-np.multiply(y,np.sin(np.divide(np.pi,4)))
rotated_y = np.multiply(x,np.sin(np.divide(np.pi,4)))+np.multiply(y,np.cos(np.divide(np.pi,4)))

plt.axis('equal')
plt.plot(rotated_x, rotated_y, 'k.',label = '1-norm unit ball')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

Using rejection sampling, draw samples from the closed unit ball for p = 0.7 using the samples drawn from unit 1-norm ball;

```python
x_list = []
y_list = []

count = 0.0
N=2500.0

for j in range(int(N)):
    x = random.random()*np.sqrt(2)-np.divide(1,np.sqrt(2))
    y = random.random()*np.sqrt(2)-np.divide(1,np.sqrt(2))
    rotated_x = np.multiply(x,np.cos(np.divide(np.pi,4)))-np.multiply(y,np.sin(np.divide(np.pi,4)))
    rotated_y = np.multiply(x,np.sin(np.divide(np.pi,4)))+np.multiply(y,np.cos(np.divide(np.pi,4)))
    
    check = np.power(np.power(np.absolute(rotated_x),0.7)+np.power(np.absolute(rotated_y),0.7),np.divide(1.0,0.7))
    
    if check <= 1:
        count+=1.0
        x_list.append(rotated_x) 
        y_list.append(rotated_y) 

plt.axis('equal')
plt.plot(x_list, y_list, 'k.',label = 'p = '+str(0.7))
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
print 'Acceptance Rate = ',np.divide(count,N)
```
